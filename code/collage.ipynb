```python
from PIL import Image, ImageOps

# Open the images
image1 = Image.open("pepper WGAN.png")
image2 = Image.open("tomato WGAN.png")
image3 = Image.open("potato WGAN.png")
image4 = Image.open("cherry WGAN.png")
image5 = Image.open("strawberry WGAN.png")

# Convert images to RGB mode
image1 = image1.convert("RGB")
image2 = image2.convert("RGB")
image3 = image3.convert("RGB")
image4 = image4.convert("RGB")
image5 = image5.convert("RGB")

# Resize the images with anti-aliasing
image1 = ImageOps.fit(image1, (800, 800), Image.ANTIALIAS)
image2 = ImageOps.fit(image2, (800, 800), Image.ANTIALIAS)
image3 = ImageOps.fit(image3, (800, 800), Image.ANTIALIAS)
image4 = ImageOps.fit(image4, (800, 800), Image.ANTIALIAS)
image5 = ImageOps.fit(image5, (800, 800), Image.ANTIALIAS)

# Adjust brightness and contrast
image1 = ImageOps.autocontrast(image1)
image2 = ImageOps.autocontrast(image2)
image3 = ImageOps.autocontrast(image3)
image4 = ImageOps.autocontrast(image4)
image5 = ImageOps.autocontrast(image5)

# Add a border
image1 = ImageOps.expand(image1, border=10, fill='white')
image2 = ImageOps.expand(image2, border=10, fill='white')
image3 = ImageOps.expand(image3, border=10, fill='white')
image4 = ImageOps.expand(image4, border=10, fill='white')
image5 = ImageOps.expand(image5, border=10, fill='white')

# Create a new image with the dimensions of the collage
collage = Image.new('RGB', (4200, 800), color='white')

# Paste the images onto the collage
collage.paste(image1, (0, 0))
collage.paste(image2, (800, 0))
collage.paste(image3, (1600, 0))
collage.paste(image4, (2400, 0))
collage.paste(image5, (3200, 0))

# Save the collage
collage.save("RF_conf_matrix_WGAN.jpg")
```
